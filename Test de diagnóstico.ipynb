1. Los top 10 tweets más retweeted
(Referencia: https://www.geeksforgeeks.org/python-pandas-dataframe-nlargest/)

```python
def top_tweets(cantidad):
    top = df.nlargest(cantidad, 'retweetCount', keep='first')
    return top
top_tweets(10)
```

2. Los top 10 usuarios en función a la cantidad de tweets que emitieron.

```python
def users_list(): #función para obtener todos los usernames
    users = []
    for user in df["user"]:
        u = user["username"]
        users.append(u)
    return users # lista con todos los nombres de usuario
```

```python
def new_column(data): #función para agregar una nueva columna al dataframe
    df['username']= data
users = users_list()
new_column(users) #agregamos la nueva columna
```

```python
def top_users(cantidad):
    df_groupby = df.groupby(['username']).count()
    top = df_groupby.nlargest(cantidad, 'content', keep='first')
    return top
top_users(10)
```

3. Los top 10 días donde hay más tweets.

```python
def top_days(cantidad):
    df_groupby = df.groupby(['date']).count()
    top = df_groupby.nlargest(cantidad, 'id', keep='first')
    return top
top_days(10)
```

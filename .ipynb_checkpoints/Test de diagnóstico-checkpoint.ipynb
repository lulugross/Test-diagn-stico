1. Los top 10 tweets más retweeted
(Referencia: https://www.geeksforgeeks.org/python-pandas-dataframe-nlargest/)

```python
def top_tweets(cantidad):
    top = df.nlargest(cantidad, 'retweetCount', keep='first')
    return top
```

2. Los top 10 usuarios en función a la cantidad de tweets que emitieron (Referencia: https://www.kite.com/blog/python/pandas-groupby-count-value-count/).

```python
def users_list(): #función para obtener todos los usernames
    users = []
    for user in df["user"]:
        u = user["username"]
        users.append(u)
    return users # lista con todos los nombres de usuario
```

```python
def new_column(data): #función para agregar una nueva columna al dataframe
    df['username']= data
users = users_list()
new_column(users) #agregamos la nueva columna
```

```python
def top_users(cantidad):
    df_groupby = df.groupby(['username']).count()
    top = df_groupby.nlargest(cantidad, 'content', keep='first')
    return top
```

3. Los top 10 días donde hay más tweets (Referencia: https://www.kite.com/blog/python/pandas-groupby-count-value-count/).

```python
def top_days(cantidad):
    df_groupby = df.groupby(['date']).count()
    top = df_groupby.nlargest(cantidad, 'id', keep='first')
    return top
```

4. Top 10 hashtags más usados (Referencia: https://methods.sagepub.com/dataset/howtoguide/regular-expressions-in-isis-twitter-2016-python#:~:text=Hashtags%20are%20strings%20of%20characters,appear%20one%20or%20more%20times).

```python
def top_hashtag(cantidad):
    hashtags=df['content'].str.extractall(r"(#\S+)")
    order = hashtags[0].value_counts()
    return order[:cantidad]
```

Funcion main

```python
def main(n):
    
    #top 10 tweets más retweeted
    if n == 1:
        t = top_tweets(10)
        return t

    #top 10 usuarios en función a la cantidad de tweets que emitieron
    if n == 2:
        u = top_users(10)
        return u
    
    #top 10 días donde hay más tweets
    if n == 3:
        d = top_days(10)
        return d

    #top 0 hashtags más usados
    if n == 4:
        h = top_hashtag(10)
        return h
```
